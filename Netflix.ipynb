```python
## Create a new column Content_Age:

     # New → Released after 2020
    #  Recent → 2015–2020
    #  Old → Before 2015 


def age(x):
    if x>2020:
        return "new"

    elif x>=2015:
        return "recent"
    else:
        return "old"
    

df["content_age"]=df["release_year"].apply(age)
```
